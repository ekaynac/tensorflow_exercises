```python
import tensorflow as tf

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import normalize
from sklearn.metrics import r2_score
```

```python
house = pd.read_csv("house_features_for_model.csv")
price = pd.read_csv("house_prices_for_model.csv")

X= house
Y= price

x_train, x_test, y_train,y_test = train_test_split(X,Y)

x_train = normalize(x_train)
x_test = normalize(x_test)
```

```python
model = tf.keras.Sequential()

girdi_adedi = x_train.shape[1]

model.add(tf.keras.layers.Dense(20, activation='relu', input_shape=(girdi_adedi,), name = "Gizli_Katman_1"))

model.add(tf.keras.layers.Dense(10, activation='relu', name = "Gizli_Katman_2"))

model.add(tf.keras.layers.Dense(1, name = "Cikti"))

model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.1),
              loss='mse',
              metrics=["mae"])
```

```python
model.fit(x_train, y_train, epochs=100)
```

```python
r2_score(y_test , model.predict(x_test))
```

```python
r2_score(y_train , model.predict(x_train))
```
